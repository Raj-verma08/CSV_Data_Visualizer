```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# function to load data from CSV file and if file was not found then print error.
def load_data(file_path):
    try:
        data = pd.read_csv(file_path)
        print(f"Data loaded successfully! Here are the first few rows:\n{data.head()}")
        return data
    except FileNotFoundError:
        print("The file was not found.")
        return None

data=load_data("real_estate_price_size.csv")
```

```python
#Line Plot 
plt.figure(figsize=(7, 5))
plt.plot(data['size'],data['price'] , marker='o', color='b', label=f"{'price'} over {'size'}")
plt.title(f"Line Plot: {'price'} vs {'size'}")
plt.xlabel('size')
plt.ylabel('price')
plt.legend()
plt.grid()
plt.show()
#plt.title("Line Plot: price vs size")
```

```python
#Box Plot:
sns.boxplot(data)
plt.title("Boxplot of Sample Data")
plt.grid()
plt.show()
```

```python
#Scatter plot:
plt.scatter(data['size'], data['price'], color='g', label=f"{'price'} over {'size'}")
plt.title(f"Scatter Plot: {'price'} vs {'size'}")
plt.xlabel('size')
plt.ylabel('price')
plt.grid()
plt.show()
```

```python
#Histogram:
plt.hist(data['price'], bins=10, color='purple', label=f"Distribution of {'price'}")
plt.title(f"Histogram of {'price'}")
plt.xlabel('price')
plt.grid()
plt.show()
```
